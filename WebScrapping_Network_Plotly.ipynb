```python
import requests

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36'}

r = requests.get('https://us.megabus.com/journey-planner/api/origin-cities',headers=headers)
```

```python
r.status_code
```

```python
org = r.json()
```

```python
org
```

```python
# Looping through the cities to get destinations from the specific city 
for x in org.get('cities'):
    params = {'originCityId':x.get('id')}
    dest = requests.get('https://us.megabus.com/journey-planner/api/destination-cities',params = params,headers=headers)
    x['dest'] = dest.json().get('cities')
```

```python
import plotly.graph_objects as go
import pandas as pd
#converting the json to Dataframe (plotly library takes data as a Dataframe)
df = pd.DataFrame(org.get('cities'))
```

```python
df
```

```python
#Map showing all the destinations
fig = go.Figure()

fig.add_trace(go.Scattergeo(
    locationmode = 'USA-states',
    lon = df['longitude'],
    lat = df['latitude'],
    hoverinfo = 'text',
    text = df['name'],
    mode = 'markers',
    showlegend = False,
    marker = dict(
        size = 5,
        color = 'orange',
        line = dict(
            width = 3,
            color = 'rgba(68, 68, 68, 0)'
        )
    )))


fig.update_layout(
    title_text = "<a href='https://us.megabus.com/'>Mega Bus</a> Destinations",
    geo = go.layout.Geo(
        scope = 'usa',
        showland = True,
        landcolor = 'rgb(243, 243, 243)',
        countrycolor = 'rgb(204, 204, 204)',
        
    ),
)

fig.show()
```

```python
for i in org.get('cities'):
    for dest in i.get('dest'):
        fig.add_trace(
            go.Scattergeo(
                locationmode = 'USA-states',
                lon = [i.get('longitude'), dest.get('longitude')],
                lat = [i.get('latitude'), dest.get('latitude')],
                mode = 'lines',
                line = dict(width=0.5,color = 'red'),
                opacity = 0.20,
                showlegend = False,
                hoverinfo='skip'
            )
        )
fig.update_layout(
    title_text = "<a href='https://us.megabus.com/'>Mega Bus</a> Routes"
        )
fig.show()
```
